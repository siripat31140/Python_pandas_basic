## Import Data From CSV File

```python
ds = pd.read_csv('data.csv')     #สามารถระบุ sheet ได้ว่าจะเอาชีทไหน
ds.head()
```

## Data Condition  (สภาพข้อมูล)

```python
ds[ds['drv'] == '4'].head()   #อยากรู้เเค่ drv ที่มีเเต่4
```

```python
ds [ds['hwy'] == ds['hwy'].max()]  #หาค่าที่มากที่สุดใน hwy เเล้วเเสดงเป็น dataframe
```

## Data Missing Value

```python
new_ds = ds.fillna( 'ffill' )   #ใส่ค่าเเทน Na   'ffill' คือเอาค่าก่อนน่ามาใส่ 
new_ds
```

## Interpolate , dropna (คือการประมาณค่า)

```python
new_dt = new_ds.interpolate()   #คือการเอาค่าประมาณไปเเทนที่ nan เเต่ไม่เเทนกลุ่มข้อมูล category
new_dt
```

```python
new_dt.iloc[0,2]  #locจะเเสดงเเค่เเถวที่สนใจ ส่วน ilocจะเเสดงตำเเหน่ง
```

```python
new_dd = new_dt.dropna()   #คือการลบเเถวที่มี nan ออกไป
new_dd
```
